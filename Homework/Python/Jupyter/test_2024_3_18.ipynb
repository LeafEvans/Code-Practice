# 偏微分方程的符号解和数值解

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
dt = 0.01
t0 = -10.0
y0 = 1
T = 10.0

t_list = [t0]
y_list = [y0]

t = t0
y = y0

while t < T:
    t += dt
    y = y + dt * (np.sin(t**2) + np.sin((t + dt) ** 2)) * 0.5
    t_list.append(t)
    y_list.append(y)

plt.plot(t_list, y_list, "c", label="T")
plt.show()
```

```python
X = 10.5
x0 = 0
y0 = 0
dx = 0.001

x = x0
y = y0

x_list = [x0]
y_list = [y0]

while x < X:
    # 先求一个近似的 y_(n+1) 以便于后面的计算求出更精确的
    y_ = y + 1 / (x**2 + 1) - 2 * y**2
    x_ = x + dx
    y = 0.5 * (1 / (x_**2 + 1) - 2 * y_**2 + 1 / (x**2 + 1) - 2 * y**2) + y
    x = x_
    y_list.append(y)
    x_list.append(x)

plt.plot(x_list, y_list, label="y(x)")
plt.legend()
plt.grid("on")
plt.show()
```

```python
dy = lambda y, x: 1 / (1 + x**2) - 2 * y**2
# 从 0 开始，每次增加 0.1，加到 10.5 为止（取不到 10.5）
x = np.arange(0, 10.5, 0.1)
# 微分方程 dy，y 的首项（y(0) 等于多少），自变量列表
sol = odeint(dy, 0, x)
print("x = {}\ny = {}".format(x, sol.T))
plt.plot(x, sol)
plt.show()
```

```python
x0 = 0
X = 0.25
dx = 0.01
y0 = 0
yp0 = 2

x_list = [x0]
y_list = [y0]
yp_list = [yp0]

x = x0
y = y0
yp = yp0

while x < X:
    y_ = y + dx * (yp)
    yp_ = yp + dx * (20 * (1 - y**2) * yp - y)
    x += dx
    y = y_
    yp = yp_
    x_list.append(x)
    y_list.append(y)
    yp_list.append(yp)

plt.plot(x_list, y_list, label="y")
plt.plot(x_list, yp_list, label="y'")
plt.legend()
plt.grid("on")
```

```python
n = 1000
dx = 0.01

x = np.zeros(n)
y = np.zeros(n)
yp = np.zeros(n)
ypp = np.zeros(n)

x[0] = 0
y[0] = 0
yp[0] = np.pi
ypp[0] = 0

for i in range(n - 1):
    x[i + 1] = 10 * i / n
    y[i + 1] = y[i] + dx * yp[i]
    yp[i + 1] = yp[i] + dx * ypp[i]
    ypp[i + 1] = ypp[i] + dx * (np.cos(x[i]) - y[i] + yp[i] - ypp[i])

plt.plot(x, y, label="y")
plt.plot(x, yp, label="y'")
plt.plot(x, ypp, label="y''")
plt.legend()
plt.grid(True)
plt.show()
```

$$
\begin{cases}
\dfrac{\partial u(x,\ t)}{\partial t} = \dfrac{\partial^2u(x,\ t)}{\partial x^2} \quad 0 \le t \le 1000,\ 0 \le x \le 1 \\
u(x,\ 0) = 4x(3-x) \\
u(0,\ t) = 0 \\
u(3,\ t) = 0
\end{cases}

```python
h = 0.1  # 空间步长
N = 30  # 空间步数
dt = 0.00001  # 时间步长，调整为更小的值
M = 100000  # 时间步数，根据新的时间步长进行调整
A = dt / h**2  # lambda*tau/h^2
U = np.zeros([N + 1, M + 1])  # 建立二维空数组
Space = np.arange(0, (N + 1) * h, h)  # 建立空间等差数列

for k in np.arange(0, M + 1):
    U[0, k] = 0.0
    U[N, k] = 0.0

for i in np.arange(0, N + 1):
    U[i, 0] = 4 * i * h * (3 - i * h)

for k in np.arange(0, M):
    for i in np.arange(1, N):
        U[i, k + 1] = U[i, k] + A * (U[i - 1, k] - 2 * U[i, k] + U[i + 1, k])

plt.plot(Space, U[:, 0], "g-", label="t = 0", linewidth=1.0)
plt.plot(Space, U[:, 30000], "b-", label="t = 3 / 10", linewidth=1.0)
plt.plot(Space, U[:, 60000], "k-", label="t = 6 / 10", linewidth=1.0)
plt.plot(Space, U[:, 90000], "r-", label="t = 9 / 10", linewidth=1.0)
plt.plot(Space, U[:, 100000], "y-", label="t = 1", linewidth=1.0)
plt.ylabel("u(x, t)", fontsize=20)
plt.xlabel("x", fontsize=20)
plt.xlim(0, 3)
plt.ylim(-2, 10)
plt.legend(loc="upper right")
plt.show()
```

```python
# 温度等高线随时空坐标的变化，温度越高，颜色越偏红
extent = [0, 1, 0, 3]  # 时间和空间的取值范围
levels = np.arange(0, 10, 0.1)  # 温度等高线的变化范围 0-10，变化间隔为 0.1
plt.contourf(U, levels, origin="lower", extent=extent, cmap=plt.cm.jet)
plt.ylabel("x", fontsize=20)
plt.xlabel("t", fontsize=20)
plt.show()
```
