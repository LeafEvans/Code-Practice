```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['font.sans-serif'] = ['FangSong']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.arange(-np.pi * 10, 10 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y, c='b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
